```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Define circles for AI, ML, and DL
ai_circle = Circle((0, 0), 3, color='skyblue', alpha=0.5, label='Artificial Intelligence (AI)')
ml_circle = Circle((0, 0), 2, color='lightgreen', alpha=0.5, label='Machine Learning (ML)')
dl_circle = Circle((0, 0), 1, color='lightcoral', alpha=0.5, label='Deep Learning (DL)')

# Add circles to the plot
ax.add_patch(ai_circle)
ax.add_patch(ml_circle)
ax.add_patch(dl_circle)

# Add labels at the center of each circle
ax.text(0, 0, 'DL', fontsize=14, ha='center', va='center')
ax.text(0, 0.8, 'ML', fontsize=16, ha='center', va='center')
ax.text(0, 1.8, 'AI', fontsize=18, ha='center', va='center')

# Remove axes for clarity
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)
ax.axis('off')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import matplotlib.patches as mpatches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Define circles with positions and sizes to represent overlaps
# AI Circle
ai_circle = Circle((0, 0), 5, color='lightgray', alpha=0.5, label='Artificial Intelligence (AI)')

# ML Circle
ml_circle = Circle((-1, 1), 3.5, color='lightgreen', alpha=0.5, label='Machine Learning (ML)')

# DL Circle inside ML
dl_circle = Circle((-1.5, 1.5), 2, color='lightcoral', alpha=0.5, label='Deep Learning (DL)')

# NLP Circle overlapping with ML and DL
nlp_circle = Circle((1, 1.5), 2.5, color='skyblue', alpha=0.5, label='Natural Language Processing (NLP)')

# Computer Vision Circle overlapping with ML and DL
cv_circle = Circle((-1, -1.5), 2.5, color='orange', alpha=0.5, label='Computer Vision (CV)')

# Robotics Circle overlapping with AI and ML
robotics_circle = Circle((2, -1), 2.5, color='purple', alpha=0.5, label='Robotics')

# Expert Systems Circle inside AI but outside ML
es_circle = Circle((0, -3), 1.5, color='yellow', alpha=0.5, label='Expert Systems (ES)')

# Add circles to the plot
for circle in [ai_circle, ml_circle, dl_circle, nlp_circle, cv_circle, robotics_circle, es_circle]:
    ax.add_patch(circle)

# Add labels near the circles
ax.text(0, 0, 'AI', fontsize=20, ha='center', va='center')
ax.text(-1, 1, 'ML', fontsize=16, ha='center', va='center')
ax.text(-1.5, 1.5, 'DL', fontsize=14, ha='center', va='center')
ax.text(1, 1.5, 'NLP', fontsize=14, ha='center', va='center')
ax.text(-1, -1.5, 'CV', fontsize=14, ha='center', va='center')
ax.text(2, -1, 'Robotics', fontsize=14, ha='center', va='center')
ax.text(0, -3, 'ES', fontsize=14, ha='center', va='center')

# Set limits and remove axes
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.axis('off')

# Create custom legend
legend_patches = [
    mpatches.Patch(color='lightgray', alpha=0.5, label='Artificial Intelligence (AI)'),
    mpatches.Patch(color='lightgreen', alpha=0.5, label='Machine Learning (ML)'),
    mpatches.Patch(color='lightcoral', alpha=0.5, label='Deep Learning (DL)'),
    mpatches.Patch(color='skyblue', alpha=0.5, label='Natural Language Processing (NLP)'),
    mpatches.Patch(color='orange', alpha=0.5, label='Computer Vision (CV)'),
    mpatches.Patch(color='purple', alpha=0.5, label='Robotics'),
    mpatches.Patch(color='yellow', alpha=0.5, label='Expert Systems (ES)'),
]

ax.legend(handles=legend_patches, loc='upper right', bbox_to_anchor=(1.2, 1))

# Display the plot
plt.title('AI Ecosystem Venn Diagram', fontsize=22)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import matplotlib.patches as mpatches
from matplotlib.animation import FuncAnimation

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Define circles with positions and sizes to represent overlaps
circles_info = [
    # (Circle object, label text, x position, y position, font size)
    (Circle((0, 0), 5, color='lightgray', alpha=0.5), 'AI', 0, 0, 20),
    (Circle((-1, 1), 3.5, color='lightgreen', alpha=0.5), 'ML', -1, 1, 16),
    (Circle((-1.5, 1.5), 2, color='lightcoral', alpha=0.5), 'DL', -1.5, 1.5, 14),
    (Circle((1, 1.5), 2.5, color='skyblue', alpha=0.5), 'NLP', 1, 1.5, 14),
    (Circle((-1, -1.5), 2.5, color='orange', alpha=0.5), 'CV', -1, -1.5, 14),
    (Circle((2, -1), 2.5, color='purple', alpha=0.5), 'Robotics', 2, -1, 14),
    (Circle((0, -3), 1.5, color='yellow', alpha=0.5), 'ES', 0, -3, 14),
]

# Create custom legend patches
legend_patches = [
    mpatches.Patch(color='lightgray', alpha=0.5, label='Artificial Intelligence (AI)'),
    mpatches.Patch(color='lightgreen', alpha=0.5, label='Machine Learning (ML)'),
    mpatches.Patch(color='lightcoral', alpha=0.5, label='Deep Learning (DL)'),
    mpatches.Patch(color='skyblue', alpha=0.5, label='Natural Language Processing (NLP)'),
    mpatches.Patch(color='orange', alpha=0.5, label='Computer Vision (CV)'),
    mpatches.Patch(color='purple', alpha=0.5, label='Robotics'),
    mpatches.Patch(color='yellow', alpha=0.5, label='Expert Systems (ES)'),
]

# Initialize empty lists to store the drawn circles and texts
drawn_circles = []
drawn_texts = []

# Function to initialize the animation
def init():
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.axis('off')
    plt.title('AI Ecosystem Venn Diagram', fontsize=22)
    return []

# Animation function
def animate(i):
    if i < len(circles_info):
        circle, label, x_text, y_text, fontsize = circles_info[i]
        ax.add_patch(circle)
        text = ax.text(x_text, y_text, label, fontsize=fontsize, ha='center', va='center')
        drawn_circles.append(circle)
        drawn_texts.append(text)
    if i == len(circles_info):
        # Add the legend after all circles have appeared
        ax.legend(handles=legend_patches, loc='upper right', bbox_to_anchor=(1.2, 1))
    return drawn_circles + drawn_texts

# Create the animation
ani = FuncAnimation(fig, animate, init_func=init, frames=len(circles_info)+1, interval=1500, blit=True, repeat=False)

# Save the animation as an MP4 video file (optional)
# ani.save('ai_ecosystem_animation.mp4', writer='ffmpeg', fps=1)

# Display the animation
plt.show()
```
